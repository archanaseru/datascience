```python
# each list should be  column in a dataframe.
id = [101, 102, 103, 104, 105, 106]
name = ['amar', 'amala', 'siva', 'sivani', 'mani', 'manisha']
age = [25, 26, 30, 23, 24, 35]
salary = [10000, 30000, 60000, 90000, 100000, 90000]
gender = ['m', 'f', 'm', 'f', 'm', 'f']
dno = [11, 12, 13, 11, 12, 14]
```

```python
df['tax'] = df['salary'] * 0.1
df
```

```python
df['hra'] = df['salary'] * 0.2
df
```

```python
df['netsal'] = df['salary'] - df['tax'] + df['hra']
df
```

```python
# Conditional transformations.
# convert gender short labels into  full lables.
# m --> "Male"
# f --> "Female"

def toGender(g):
   if g.lower()=='m':
      g='Male'
   elif g.lower()=='f':
     g = 'Female'
   else:
     g='Invalid'
   return g


df['gender'] = df['gender'].apply(toGender)
df
```

```python
def  toDept(d):
  dname = 'Other'
  dept = {11:"Marketing", 12:"Hr", 13:"Finance"}
  if d in dept:
    dname = dept[d]
  return dname

df['dname']  = df['dno'].apply(toDept)
print(df)
```

```python
# salary --> grade
def toGrade(s):
   grade = 'D'
   if s>=100000:
     grade = 'A'
   elif s>=75000:
     grade='B'
   elif s>=50000:
     grade='C'
   return grade

df['grade'] = df['salary'].apply(toGrade)
print(df)
```

```python
# age --> into age groups
def ageGroup(age):
   grp = 'old'
   if age<=12:
     grp = 'Child'
   elif age<=19:
     grp = 'teen'
   elif age<=35:
     grp = 'youth'
   elif age<=50:
     grp = 'adult'
   elif age<=65:
    grp = 'middle'
   return grp

df['ageGrp'] = df['age'].apply(ageGroup)
print(df)


```
