```python
import pandas as pd
```

```python

id = [101, 102, 103, 104, 105, 106]
name = ['amar', 'amala', 'siva', 'sivani', 'mani', 'manisha']
age = [25, 26, 30, 23, 24, 35]
salary = [10000, 30000, 60000, 90000, 100000, 90000]
gender = ['m', 'f', 'm', 'f', 'm', 'f']
dno = [11, 12, 13, 11, 12, 14]
pd1= pd.DataFrame([id,name,age,salary,gender,dno])
pd1
```

```python
print(id,type(id),len(id))
print(name,type(name),len(name))
print(age,type(age),len(age))
print(salary,type(salary),len(salary))
print(gender,type(gender),len(gender))
print(dno,type(dno),len(dno))
# dataframe  each list objects as row
```
